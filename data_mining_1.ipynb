####4.
Dataframe'i 8.sütun ('date') sütununda aylara göre gruplayarak [9-14] sütunlarını aşağıdakine benzer şekilde plot ediniz (groupby ve DatetimeIndex kullanarak yapabilirsiniz).

Örnek
mobibilitybase.png




```python
import pandas as pd

date= pd.to_datetime(df['date']) #date sütununu çektim
index = pd.DatetimeIndex(date.values).month #bizim için gerekli olan aylar olduğu için sadece month'u çektim, ve bunu artık tarih indexi yerine sadece aylar index olacak
df= df.set_index(index) #index olarak df'e ayları verdik [2,4,6,8,19,12]
df.index.name = 'Aylar' #index sütunun isimin aylar olara değiştirildi
df= df.iloc[:, 9:15]

gr1 = df.groupby('Aylar').sum()
gr1.plot(figsize =(5, 4),linewidth=4.0).legend(bbox_to_anchor=(1.0, 1.0)) #aylara göre gruplama

plt.title('Aylara Göre Gruplama ve Karşılaştırma') #x eksenine isimleri değiştir
plt.show()
```
